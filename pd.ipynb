## 4- Data Filtering 

### 4.1- Column Filtering

#### 4.1.1- Column Selection 

```python
# Get a single columns you have two ways 
# first the column name is a single word that doesnot contain a white space
df.median_house_value
```

```python
# if the column name has a white space on its name put the between ['']
df['median_house_value']
```

```python
# to in get multi columns at ones you need to pass the names within a list 
df[['median_house_value', 'total_bedrooms']]
```

#### 4.1.2- Column Unique values

```python
# note this will work better on object columns 
df.ocean_proximity.unique()
```

### 4.2- Row Filtering

#### 4.2.1- Selecting row with specific value 

```python
# you can filter with one condition and also with muli condition
# single condition
df[df['ocean_proximity'] == "INLAND"]
```

```python
# multi condition
df[(df['ocean_proximity'] == 'INLAND') & (df['housing_median_age'] > 30)]
```

#### 4.2.2- Selecting with index

```python
# we will use iloc and there are another method called loc that works with keyword instead of index
# you can select a single row
df.iloc[0]
```

```python
# you can select within a range 
df.iloc[0:6]
```

```python
# you can select one value within a row by typeing its place [row no, column no.]
df.iloc[0,1]
```
