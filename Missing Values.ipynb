### filling a missing value with previous ones  

```python
# filling a missing value with previous ones  
df.fillna(method ='pad')
```

### filling a missing value with next ones  

```python
# Filling null value with the next ones 
df.fillna(method ='bfill')
```

### filling a null values using specific value  

```python
# filling a null values using specific value 
df = df["total_bedrooms"].fillna("No Bedroom", inplace = True) 
```

```python
# will replace  Nan value in dataframe with value No Bedroom  
df.replace(to_replace = np.nan, value = "No Bedroom") 
```

### Drop null  

```python
# using dropna() function  
df.dropna()
```

```python
# Now we drop a rows whose all data is missing or contain null values(NaN) 
df.dropna(how = 'all')
```

```python
# Dropping columns with at least 1 null value
df.dropna(axis = 1)
```

### filling with the mean

```python
#Finding the mean of the column having NaN
mean_value = df['total_bedrooms'].mean()
mean_value
```

```python
df['total_bedrooms'].fillna(value=mean_value, inplace=True)
```

```python
# To calculate mean use imputer class
from sklearn.impute import SimpleImputer

new_df = df.drop('ocean_proximity', axis=1)
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit_transform(new_df)
```

```python
new_df.isnull().sum()
```

### filling with the median

```python
median_value = df['total_bedrooms'].median()
median_value
```

```python
df['total_bedrooms'].fillna(value=median_value, inplace=True)
```

### filling with the Mode

```python
mode_value = df['total_bedrooms'].mode()
mode_value
```

```python
df['total_bedrooms'].fillna(value=mode_value, inplace=True)
```

### filling with the Knn 

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
imputer.fit_transform(new_df)
```
